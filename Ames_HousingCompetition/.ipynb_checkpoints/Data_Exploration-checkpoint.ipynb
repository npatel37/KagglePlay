```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train_rawdf = pd.read_csv("train.csv")
test_rawdf = pd.read_csv("test.csv")

data_rawdf = pd.concat((train_rawdf.loc[:,'MSSubClass':'SaleCondition'],
                      test_rawdf.loc[:,'MSSubClass':'SaleCondition']), ignore_index=True)
```

```python
#print data_rawdf
```

```python
def scatterplot(xname,yname,data):
    plt.scatter(train_rawdf[xname],train_rawdf[yname])
    plt.xlabel(xname,size=18)
    plt.xticks(size=14)
    plt.yticks(size=14)
    #plt.xlim(xmin=0,xmax=50000)
    plt.show()
    plt.close()
    
yname = 'SalePrice'
scatterplot('LotArea',yname,train_rawdf)
scatterplot('YearBuilt',yname,train_rawdf)
scatterplot('YearRemodAdd',yname,train_rawdf)
scatterplot('MasVnrArea',yname,train_rawdf)
scatterplot('BsmtFinSF1',yname,train_rawdf)
scatterplot('BsmtFinSF2',yname,train_rawdf)
scatterplot('BsmtUnfSF',yname,train_rawdf)
scatterplot('TotalBsmtSF',yname,train_rawdf)
scatterplot('1stFlrSF',yname,train_rawdf)
scatterplot('2ndFlrSF',yname,train_rawdf)
scatterplot('GrLivArea',yname,train_rawdf)
scatterplot('GarageYrBlt',yname,train_rawdf)
scatterplot('GarageArea',yname,train_rawdf)
scatterplot('WoodDeckSF',yname,train_rawdf)
scatterplot('OpenPorchSF',yname,train_rawdf)
scatterplot('EnclosedPorch',yname,train_rawdf)
scatterplot('3SsnPorch',yname,train_rawdf)
scatterplot('ScreenPorch',yname,train_rawdf)
scatterplot('PoolArea',yname,train_rawdf)
scatterplot('MiscVal',yname,train_rawdf)
scatterplot('MoSold',yname,train_rawdf)
scatterplot('YrSold',yname,train_rawdf)
```

```python
keynames = data_rawdf.keys()

for columnName in keynames:

    Ulist = pd.Series.unique(train_rawdf[columnName])
    priceList = []
    stdDevList = []
    if (len(Ulist)>26):
            continue

    for string in Ulist:
        df_string = train_rawdf[train_rawdf[columnName] == string]
        average = df_string['SalePrice'].mean()
        error = df_string['SalePrice'].std()
        priceList.append(average)
        stdDevList.append(error)

    perm = np.argsort(priceList)[::-1]
    priceList = [priceList[x] for x in perm]
    Ulist = [Ulist[x] for x in perm]
    stdDevList = [stdDevList[x] for x in perm]
    
    ax = plt.subplot(111)
    width = 0.3
    bins = map(lambda x: x-width/2,range(1,len(priceList)+1))
    ax.bar(bins,priceList,yerr=stdDevList,color='yellow',ecolor='black')
    ax.set_xticks(map(lambda x: x, range(1,len(priceList)+1)))
    ax.set_xticklabels(Ulist,rotation=90,size=18)
    ax.set_title(columnName, size=18)
    
    del Ulist, priceList, stdDevList
    plt.show()
    plt.close()
    
```
